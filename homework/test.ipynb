```python
import pandas as pd

with open("../files/input/clusters_report.txt") as file:
    clusters_report = file.read()

clusters_report = clusters_report.split('\n')
clusters_report
```

```python
columnas_1 = clusters_report[0]
columnas_2 = clusters_report[1]

columnas_aux = columnas_1.split(' ')
columnas_1 = []
nombre = ''

for i in range(len(columnas_aux)):
    if columnas_aux[i] != '':
        nombre += ' '+columnas_aux[i]
    if columnas_aux[i] == '' and i<=len(columnas_aux):
        columnas_1.append(nombre)
        nombre = ''

texto = "palabras clave"

columnas = [i.lstrip() for i in columnas_1 if i != '']
columnas[2] += ' ' + texto
columnas[1] += ' ' + texto
columnas = [columna.lower().replace(' ','_') for columna in columnas]
columnas = '/-/'.join(columnas)
columnas
```

```python
datos_sucios = clusters_report[4::]

numerador = 1
lista_aux = []
for linea in datos_sucios:
    linea = linea.strip()
    if linea.startswith(str(numerador)):
        lista_aux.append('/-/')
        lista_aux.append(linea)
        numerador += 1
    else:
        lista_aux.append(linea)

datos = ''.join(lista_aux)
datos = datos.split('/-/')

datos = [i.strip().split() for i in datos]
datos.pop(0)
datos

datos_org = []
contador = 0
for dato in datos:
        claves_aux = []
        datos_org.append([])
        datos_org[contador].append(dato[0])
        datos_org[contador].append(dato[1])
        datos_org[contador].append(dato[2].replace(',','.'))
        for palabra in range(4,len(dato)):
            claves_aux.append(dato[palabra])
        print(claves_aux)
        datos_org[contador].append(' '.join(claves_aux))
        contador += 1

clusters = ['/-/'.join(dato) for dato in datos_org]
clusters
```

```python
with open("../files/input/clusters_report_copy.txt", "w") as archivo:
    archivo.write(str(columnas) + '\n')
    for i in clusters:
        archivo.write(i.strip('.') + '\n')
```
